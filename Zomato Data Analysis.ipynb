```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df= pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
#To check Datatypes
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val =final_df.Country.value_counts().values
```

```python
## Pie Chart
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Zomato Maximum No of transection done in India after that USA then UK


```python
final_df.columns
```

```python
raiting = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Raiting Count'})
```

```python
raiting
```

1. when raiting is between 4.5 to 4.9 ---> Excellent
2. when raiting is between 4.0 to 4.4 ---> Very Good
3. when raiting is between 3.9 to 3.5 ---> Good
4. when raiting is between 3.4 to 2.5 ---> Average
5. when raiting is between 2.4 to 1.8 ---> Poor


Zero raiting also given lots of coustmer

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Raiting Count',hue='Rating color',data=raiting, palette=['blue','red','orange','yellow','green','green'])
```

1. Not Raiting counts is very high

```python
plt.pie(city_value[:5],labels=city_index[:5],autopct='%1.2f%%')
```

```python
#Find the top 10 Cuisines
cuisines_index = final_df.Cuisines.value_counts().index
cuisines_value = final_df.Cuisines.value_counts().values
```

```python
plt.pie(cuisines_value[:10],labels=cuisines_index[:10],autopct='%1.2f%%')
```

```python
#Finding top 10 restaurant
res_index = final_df['Restaurant Name'].value_counts().index
res_value = final_df['Restaurant Name'].value_counts().values
```
